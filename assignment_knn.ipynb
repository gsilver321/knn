**Q1.** This question is a case study for $k$ nearest neighbor The target variable `y` is `price` and the features are `year` and `mileage`.

1. Load the `./data/USA_cars_datasets.csv`. Keep the following variables and drop the rest: `price`, `year`, `mileage`. Are there any `NA`'s to handle? Look at the head and dimensions of the data.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

! git clone https://github.com/DS3001/knn
````

````python
df = pd.read_csv('knn/data/USA_cars_datasets.csv')
df = df.loc[:, ['price','year','mileage'] ]

print("NA Count:")
for c in df.columns:
  print(c, sum(df[c].isnull()))

# no NA's to handle ^^

print('\n', df.shape)
df.head()
````

2. Maxmin normalize `year` and `mileage`.

````python
def maxmin(x):
    return (x-min(x))/(max(x)-min(x))

df['year'] = maxmin(df['year'])
df['mileage'] = maxmin(df['mileage'])

df.head()
````

3. Split the sample into ~80% for training and ~20% for evaluation.

````python
X_train, X_test, y_train, y_test = train_test_split(df.drop('price', axis=1), df['price'], test_size=0.2, random_state=100)
````

4. Use the $k$NN algorithm and the training data to predict `price` using `year` and `mileage` for the test set for $k=3,10,25,50,100,300$. For each value of $k$, compute the mean squared error and print a scatterplot showing the test value plotted against the predicted value. What patterns do you notice as you increase $k$?

````python
for k in [3, 10, 25, 50, 100, 300]:
    mdl = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
    y_hat = mdl.predict(X_test)
    SSE = np.sum((y_test - y_hat)**2)

    plot, ax = plt.subplots()

    ax.set_ylim(-1000, 62000)
    ax.set_xlim(-1000, 62000)

    plt.scatter(y_test, y_hat)
    plt.xlabel("actual")
    plt.ylabel('predicted')
    plt.title('k: ' + str(k) + ', SSE: ' + str(SSE))
    plt.show()
````

As k increases, the max and min y values of the points shrinks closer to an average value in the middle of the data. Additionally, the points generally become more clustered with less outliers

5. Determine the optimal $k$ for these data.

````python
def optimal_k(bar):
  k_bar = bar
  k_arr = np.arange(1, k_bar)
  SSE = np.zeros(k_bar)

  for k in range(k_bar):
      mdl_fit = KNeighborsRegressor(n_neighbors=k+1).fit(X_train, y_train)
      y_hat = mdl_fit.predict(X_test)
      SSE[k] = np.sum((y_test - y_hat)**2)

  opt_k = k_arr[np.where(SSE == np.min(SSE))]
  print("Optimal k:", opt_k)

  plt.plot(np.arange(0, k_bar), SSE)

  plt.xlabel("k")
  plt.ylabel('SSE')
  plt.title("k vs SSE | Optimal k: " + str(opt_k) + " | SSE: " + str(np.min(SSE)))

  plt.show()

optimal_k(200)
````

6. Describe what happened in the plots of predicted versus actual prices as $k$ varied, taking your answer into part 6 into account. (Hint: Use the words "underfitting" and "overfitting".)

In the plots that used low k values like 3, 10, and 25, the predictions varied a lot causing multiple inaccuracies (as seen with the high SSEs). With the optimal k for this model being 77, the k values of 3, 10, and 25 were likely underfitting the model. On the opposite end of the spectrum, the k value of 300 was probably overfitting the model since the data became clustered towards some average value with an even distribution. However k values like 50 and 100 that were close enough to the optimal k did a better job of predicting the prices (as seen with the lower SSEs)
